```python
import tkinter as tk
from tkinter import ttk, messagebox
import time
from plyer import notification

valid_users = {"user1": "password1", "user2": "password2"}

maintenance_records = []

# Function to log in
def login():
    username = username_entry.get()
    password = password_entry.get()
    
    if username in valid_users and valid_users[username] == password:
        show_main_menu()
    else:
        messagebox.showerror("Login Error", "Invalid username or password")

# Function to show the main menu
def show_main_menu():
    login_frame.grid_remove()
    main_menu_frame.grid()
    load_maintenance_records()

# Function to add a maintenance record
def add_maintenance():
    date = date_entry.get()
    maintenance_type = maintenance_type_entry.get()
    cost = cost_entry.get()
    notes = notes_entry.get("1.0", "end-1c")

    maintenance_records.append({"Date": date, "Type": maintenance_type, "Cost": cost, "Notes": notes})
    update_maintenance_listbox()

    date_entry.delete(0, tk.END)
    maintenance_type_entry.delete(0, tk.END)
    cost_entry.delete(0, tk.END)
    notes_entry.delete("1.0", tk.END)

# Function to load maintenance records
def load_maintenance_records():
    maintenance_listbox.delete(0, tk.END)
    for record in maintenance_records:
        maintenance_listbox.insert(tk.END, f"{record['Date']} - {record['Type']} - ${record['Cost']}")

# Function to update the maintenance listbox
def update_maintenance_listbox():
    load_maintenance_records()

# Function to delete a selected maintenance record
def delete_maintenance():
    selected_item = maintenance_listbox.curselection()
    if selected_item:
        index = selected_item[0]
        maintenance_records.pop(index)
        update_maintenance_listbox()

# Function to set a reminder
def set_reminder():
    reminder_time = time_entry.get()  # Get the reminder time in seconds
    time.sleep(int(reminder_time))
    notification_title = "Car Maintenance Reminder"
    notification_text = "It's time for your scheduled maintenance!"
    notification.notify(
        title=notification_title,
        message=notification_text,
        timeout=10  # Display notification for 10 seconds
    )

# Function to search records
def search_records():
    keyword = search_entry.get().lower()
    matching_records = [record for record in maintenance_records if keyword in record["Type"].lower()]
    
    # Clear the listbox
    maintenance_listbox.delete(0, tk.END)
    
    # Display matching records
    for record in matching_records:
        maintenance_listbox.insert(tk.END, f"{record['Date']} - {record['Type']} - ${record['Cost']}")

# Create the main application window
app = tk.Tk()
app.title("Car Maintenance Tracker")

login_frame = ttk.Frame(app)
main_menu_frame = ttk.Frame(app)

login_frame.grid(row=0, column=0, padx=50, pady=50)
main_menu_frame.grid(row=0, column=0, padx=50, pady=50)
main_menu_frame.grid_remove()

# Login Page
ttk.Label(login_frame, text="Username:").grid(row=0, column=0)
username_entry = ttk.Entry(login_frame)
username_entry.grid(row=0, column=1)

ttk.Label(login_frame, text="Password:").grid(row=1, column=0)
password_entry = ttk.Entry(login_frame, show="*")
password_entry.grid(row=1, column=1)

login_button = ttk.Button(login_frame, text="Login", command=login)
login_button.grid(row=2, columnspan=2)

# Main Menu
ttk.Label(main_menu_frame, text="Car Maintenance Tracker").grid(row=0, column=0, columnspan=2)

ttk.Label(main_menu_frame, text="Date:").grid(row=1, column=0)
date_entry = ttk.Entry(main_menu_frame)
date_entry.grid(row=1, column=1)

ttk.Label(main_menu_frame, text="Maintenance Type:").grid(row=2, column=0)
maintenance_type_entry = ttk.Entry(main_menu_frame)
maintenance_type_entry.grid(row=2, column=1)

ttk.Label(main_menu_frame, text="Cost:").grid(row=3, column=0)
cost_entry = ttk.Entry(main_menu_frame)
cost_entry.grid(row=3, column=1)

ttk.Label(main_menu_frame, text="Notes:").grid(row=4, column=0)
notes_entry = tk.Text(main_menu_frame, height=5, width=30)
notes_entry.grid(row=4, column=1)

add_button = ttk.Button(main_menu_frame, text="Add Maintenance Record", command=add_maintenance)
add_button.grid(row=5, column=0, columnspan=2)

delete_button = ttk.Button(main_menu_frame, text="Delete Selected Record", command=delete_maintenance)
delete_button.grid(row=6, column=0, columnspan=2)

maintenance_listbox = tk.Listbox(main_menu_frame, height=10, width=50)
maintenance_listbox.grid(row=7, column=0, columnspan=2)

# Reminder Section
ttk.Label(main_menu_frame, text="Set Reminder (in seconds):").grid(row=8, column=0)
time_entry = ttk.Entry(main_menu_frame)
time_entry.grid(row=8, column=1)
set_reminder_button = ttk.Button(main_menu_frame, text="Set Reminder", command=set_reminder)
set_reminder_button.grid(row=8, column=2)

# Search Section
ttk.Label(main_menu_frame, text="Search Maintenance Records:").grid(row=9, column=0)
search_entry = ttk.Entry(main_menu_frame)
search_entry.grid(row=9, column=1)
search_button = ttk.Button(main_menu_frame, text="Search", command=search_records)
search_button.grid(row=9, column=2)

app.mainloop()
```
